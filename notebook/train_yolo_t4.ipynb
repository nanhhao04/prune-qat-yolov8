```python
import yaml
import os

# Đường dẫn đến file data.yaml bên trong dự án
data_yaml_path = '/content/yolo_pipeline_hao/configs/data.yaml'

with open(data_yaml_path, 'r') as f:
    data = yaml.safe_load(f)

# SỬA LẠI ĐƯỜNG DẪN CHÍNH XÁC CHO COLAB
# Chú ý có thêm 'yolo_pipeline_hao' trong đường dẫn
data['path'] = '/content/yolo_pipeline_hao/datasets/traffic'
data['train'] = 'images/train'
data['val'] = 'images/val'
data['test'] = 'images/val'

with open(data_yaml_path, 'w') as f:
    yaml.dump(data, f)

print("✅ Đã cập nhật lại data.yaml chính xác!")
# Kiểm tra xem thư mục có tồn tại không
if os.path.exists(data['path']):
    print(f"Thư mục dữ liệu tồn tại: {data['path']}")
else:
    print(f"⚠️ CẢNH BÁO: Thư mục {data['path']} không tìm thấy!")
```
